```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
T_min_t = 21
T_max_t = 24
T = np.arange(12.5, 32.5, 0.05)

def reward(T):
    T_comfort = -1 * (1-math.exp(-1.0/2.0*(T - 22.5)**2) + 0.2*(max(0, (T - T_max_t)) + max(0, (T_min_t - T))))
    return T_comfort


reward2 = np.vectorize(reward)
```

```python
# Create the plot
plt.plot(T,reward2(T),color='gray',linewidth=4)
plt.axvline(x=21,linewidth=4)
plt.axvline(x=24,linewidth=4)

# Show the plot
plt.show()
```
